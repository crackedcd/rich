```python
import pandas as pd
```

```python
df = pd.read_csv(r"./data2.csv", encoding="gbk")
```

```python
df["day"] = time_value.day
```

```python
df.loc[:, "month"] = time_value.month
df.loc[:, "hour"] = time_value.hour
df.loc[:, "minute"] = time_value.minute
```

```python
df.groupby("receipt_address").nunique()["receipt_phone"].reset_index().query("receipt_phone > 1")
```

```python
df_new = df[["month", "day", "hour", "minute", "final_score"]]
```

```python
df_new
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
```

```python
df_final = df_new
```

```python
data = df_final[["month", "day", "hour", "minute"]]
target = df_final["final_score"]
```

```python
data_train, data_test, target_train, target_test = train_test_split(data, target)
```

```python
ss = StandardScaler()
data_train = ss.fit_transform(data_train)
data_test = ss.transform(data_test)

k = KNeighborsClassifier()
neighbor_dict = {"n_neighbors": [3, 4, 5, 6, 7, 8, 9, 10]}
gs = GridSearchCV(k, param_grid=neighbor_dict, cv=30)
gs.fit(data_train, target_train)
target_predict = gs.predict(data_test)
print("最佳参数")
print(gs.best_params_)
print("最佳结果")
print(gs.best_score_)
print("最佳预估器")
print(gs.best_estimator_)
print("交叉验证结果")
print(gs.cv_results_)
```
